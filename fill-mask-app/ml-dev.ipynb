```python

from transformers import pipeline
```

```python
senti_analysis_pipe = pipeline("sentiment-analysis")
senti_analysis_pipe("I've been waiting for a HuggingFace course my whole life.")
```

```python
senti_analysis_pipe(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

```python
input_sentence = input("Enter a sentance: ")
senti_analysis_pipe(input_sentence)
```

```python
unmasker = pipeline('fill-mask')

input_sentence = "Mama's don't let your <mask> grow up to be cowboys"
unmasker(input_sentence, top_k = 2)
```

```python
unmasker = pipeline('feature-extraction')

input_sentence = "Mama's don't let your <mask> grow up to be cowboys"
unmasker(input_sentence)
```

```python
generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```
